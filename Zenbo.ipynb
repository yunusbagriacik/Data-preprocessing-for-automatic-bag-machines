## Zenbo 430 Relations

```python
'''
'=+EĞER(VE(J144>=54;J144<=120);1;0) 'Tabaka En'   
=+EĞER(VE(I144>=32;I144<=60);1;0) 'Tabaka Boy'
=+EĞER(VE(H144>=120;H144<=250);1;0)  'Kağıt Gramajı.1'
=+EĞER(VE(G144>=3;G144<=6);1;0)  'Ağız Katlama Boy'
=+EĞER(VE((D144-1,5)>=16;(D144-1,5)<=41);1;0) 'Ağız Takviye En'
=+EĞER(VE((G144-0,5)>=2,5;(G144-0,5)<=5);1;0) 'Ağız Takviye Boy'
=+EĞER(VE(D144>=18;D144<=43);1;0)  'Çanta En'
=+EĞER(VE(E144>=8;E144<=17);1;0)  'Körük.1'
=+EĞER(VE((F144+(E144/2)+1,5)>=28;(F144+(E144/2)+1,5)<=57);1;0) 'Çanta Tüp Uzunluğu' 

J = 'Bıçak Bitmiş Ebadı En'
I = 'Bıçak Bitmiş Ebadı Boy'
H = 'Kağıt Gramajı'
G = 'Ağız'
D = 'En'
E = 'Körük'
F = 'Boy'

'''

import numpy as np

# Column assignments
J1 = 'Bıçak Bitmiş Ebadı En'
I1 = 'Bıçak Bitmiş Ebadı Boy'
H1 = 'Kağıt Gramajı'
G1 = 'Ağız'
D1 = 'En'
E1 = 'Körük'
F1 = 'Boy'

# Calculate 'Tabaka En' column
zenbo_luks_necessary_checked['Tabaka En'] = np.where((zenbo_luks_dropped1[J1] >= 54) & (zenbo_luks_dropped1[J1] <= 120), 1, 0)

# Calculate 'Tabaka Boy' column
zenbo_luks_necessary_checked['Tabaka Boy'] = np.where((zenbo_luks_dropped1[I1] >= 32) & (zenbo_luks_dropped1[I1] <= 60), 1, 0)

# Calculate 'Kağıt Gramajı.1' column
zenbo_luks_necessary_checked['Kağıt Gramajı.1'] = np.where((zenbo_luks_dropped1[H1] >= 120) & (zenbo_luks_dropped1[H1] <= 250), 1, 0)

# Calculate 'Ağız Katlama Boy' column
zenbo_luks_necessary_checked['Ağız Katlama Boy'] = np.where((zenbo_luks_dropped1[G1] >= 3) & (zenbo_luks_dropped1[G1] <= 6), 1, 0)

# Calculate 'Ağız Takviye En' column
zenbo_luks_necessary_checked['Ağız Takviye En'] = np.where(((zenbo_luks_dropped1[D1] - 1.5) >= 16) & ((zenbo_luks_dropped1[D1] - 1.5) <= 41), 1, 0)

# Calculate 'Ağız Takviye Boy' column
zenbo_luks_necessary_checked['Ağız Takviye Boy'] = np.where(((zenbo_luks_dropped1[G1] - 0.5) >= 2.5) & ((zenbo_luks_dropped1[G1] - 0.5) <= 5), 1, 0)

# Calculate 'Çanta En' column
zenbo_luks_necessary_checked['Çanta En'] = np.where((zenbo_luks_dropped1[D1] >= 18) & (zenbo_luks_dropped1[D1] <= 43), 1, 0)

# Calculate 'Körük.1' column
zenbo_luks_necessary_checked['Körük.1'] = np.where((zenbo_luks_dropped1[E1] >= 8) & (zenbo_luks_dropped1[E1] <= 17), 1, 0)

# Calculate 'Çanta Tüp Uzunluğu' column
zenbo_luks_necessary_checked['Çanta Tüp Uzunluğu'] = np.where(((zenbo_luks_dropped1[F1] + (zenbo_luks_dropped1[E1] / 2) + 1.5) >= 28) & ((zenbo_luks_dropped1[F1] + (zenbo_luks_dropped1[E1] / 2) + 1.5) <= 57), 1, 0)


# Check if any of the specified columns have a value of 0
zero_check = ((zenbo_luks_necessary_checked['Tabaka En'] == 0) |
              (zenbo_luks_necessary_checked['Tabaka Boy'] == 0) |
              (zenbo_luks_necessary_checked['Kağıt Gramajı.1'] == 0) |
              (zenbo_luks_necessary_checked['Ağız Katlama Boy'] == 0) |
              (zenbo_luks_necessary_checked['Ağız Takviye En'] == 0) |
              (zenbo_luks_necessary_checked['Ağız Takviye Boy'] == 0) |
              (zenbo_luks_necessary_checked['Çanta En'] == 0) |
              (zenbo_luks_necessary_checked['Körük.1'] == 0) |
              (zenbo_luks_necessary_checked['Çanta Tüp Uzunluğu'] == 0))

# Calculate the 'ZB 1200CT-430' column with the conditions
zenbo_luks_necessary_checked['ZB 1200CT-430'] = np.where(zero_check, 0, 1)
```

## Zenbo 240 Relations

```python
'''
    For the column'Tabaka En.1', an excel formula =+EĞER(VE(J144>=32,5;J144<=72);1;0) is calculated
    'Tabaka Boy.1', =+EĞER(VE(I144>=22;I144<=46);1;0)
    'Kağıt Gramajı.2', =+EĞER(VE(H144>=120;H144<=250);1;0)
    'Ağız Katlama Boy.1', =+EĞER(VE(G144>=3;G144<=5);1;0)
    'Ağız Takviye En.1', =+EĞER(VE((D144-1,5)>=9;(D144-1,5)<=22);1;0)
    'Ağız Takviye Boy.1',  =+EĞER(VE((G144-0,5)>=2,5;(G144-0,5)<=4);1;0)
    'Çanta En.1', =+EĞER(VE(D144>=11;D144<=24);1;0)
    'Körük.2',  =+EĞER(VE(E144>=5;E144<=12);1;0)
    'Çanta Tüp Uzunluğu.1', =+EĞER(VE((F144+(E144/2)+1,5)>=18;(F144+(E144/2)+1,5)<=43);1;0)
    
    
    J = 'Bıçak Bitmiş Ebadı En'
    I = 'Bıçak Bitmiş Ebadı Boy'
    H = 'Kağıt Gramajı'
    G = 'Ağız'
    D = 'En'
    E = 'Körük'
    F = 'Boy'
    
    
'''    

import numpy as np

# Column assignments
J1 = 'Bıçak Bitmiş Ebadı En'
I1 = 'Bıçak Bitmiş Ebadı Boy'
H1 = 'Kağıt Gramajı'
G1 = 'Ağız'
D1 = 'En'
E1 = 'Körük'
F1 = 'Boy'

zenbo_luks_necessary_checked['Tabaka En.1'] = np.where((zenbo_luks_dropped1['Bıçak Bitmiş Ebadı En'] >= 32.5) & (zenbo_luks_dropped1['Bıçak Bitmiş Ebadı En'] <= 72), 1, 0)

zenbo_luks_necessary_checked['Tabaka Boy.1'] = np.where((zenbo_luks_dropped1['Bıçak Bitmiş Ebadı Boy'] >= 22) & (zenbo_luks_dropped1['Bıçak Bitmiş Ebadı Boy'] <= 46), 1, 0)

zenbo_luks_necessary_checked['Kağıt Gramajı.2'] = np.where((zenbo_luks_dropped1['Kağıt Gramajı'] >= 120) & (zenbo_luks_dropped1['Kağıt Gramajı'] <= 250), 1, 0)

zenbo_luks_necessary_checked['Ağız Katlama Boy.1'] = np.where((zenbo_luks_dropped1['Ağız'] >= 3) & (zenbo_luks_dropped1['Ağız'] <= 5), 1, 0)

zenbo_luks_necessary_checked['Ağız Takviye En.1'] = np.where(((zenbo_luks_dropped1['En'] - 1.5) >= 9) & ((zenbo_luks_dropped1['En'] - 1.5) <= 22), 1, 0)

zenbo_luks_necessary_checked['Ağız Takviye Boy.1'] = np.where(((zenbo_luks_dropped1['Ağız'] - 0.5) >= 2.5) & ((zenbo_luks_dropped1['Ağız'] - 0.5) <= 4), 1, 0)

zenbo_luks_necessary_checked['Çanta En.1'] = np.where((zenbo_luks_dropped1['En'] >= 11) & (zenbo_luks_dropped1['En'] <= 24), 1, 0)

zenbo_luks_necessary_checked['Körük.2'] = np.where((zenbo_luks_dropped1['Körük'] >= 5) & (zenbo_luks_dropped1['Körük'] <= 12), 1, 0)

zenbo_luks_necessary_checked['Çanta Tüp Uzunluğu.1'] = np.where(((zenbo_luks_dropped1['Boy'] + (zenbo_luks_dropped1['Körük'] / 2) + 1.5) >= 18) & ((zenbo_luks_dropped1['Boy'] + (zenbo_luks_dropped1['Körük'] / 2) + 1.5) <= 43), 1, 0)


# Check if any of the specified columns have a value of 0
zero_check = ((zenbo_luks_necessary_checked['Tabaka En.1'] == 0) |
              (zenbo_luks_necessary_checked['Tabaka Boy.1'] == 0) |
              (zenbo_luks_necessary_checked['Kağıt Gramajı.2'] == 0) |
              (zenbo_luks_necessary_checked['Ağız Katlama Boy.1'] == 0) |
              (zenbo_luks_necessary_checked['Ağız Takviye En.1'] == 0) |
              (zenbo_luks_necessary_checked['Ağız Takviye Boy.1'] == 0) |
              (zenbo_luks_necessary_checked['Çanta En.1'] == 0) |
              (zenbo_luks_necessary_checked['Körük.2'] == 0) |
              (zenbo_luks_necessary_checked['Çanta Tüp Uzunluğu.1'] == 0))

# Calculate the 'ZB 1200CT-430' column with the conditions
zenbo_luks_necessary_checked['ZB 700CT-240'] = np.where(zero_check, 0, 1)
```

## Zenbo 60s(Soğuk Tutkal) Relations

```python
'''
    For the column 'Kağıt Gramajı.3', an excel formula =+EĞER(VE(H144>=120;H144<=250);1;0) is calculated
    'Çanta En.2', =+EĞER(VE(D144>=18;D144<=43);1;0)
    'Körük Açık Çanta Yüksekliği',=+EĞER(VE(((E144/2)+F144+1,5>=23);((E144/2)+F144+1,5<=50));1;0)
    'Körük.3', =+EĞER(VE(E144>=8;E144<=17);1;0)
    'Dip Takviye En', =+EĞER(VE((D144-0,5)>=17,4;(D144-0,5)<=42,4);1;0)
    'Dip Takviye Boy',  =+EĞER(VE((E144-0,5)>=7,4;(E144-0,5)<=16,4);1;0)
    'Körük Açık Ağız ve Körük Son Uç Arası', =EĞER(F144-(E144+(E144/2+1,5))>=4;1;0)
    'Körük Açık Katsız Kısım', =EĞER(D144-(2*(E144/2+1,5))>=5;1;0)
    

    
    J = 'Bıçak Bitmiş Ebadı En'
    I = 'Bıçak Bitmiş Ebadı Boy'
    H = 'Kağıt Gramajı'
    G = 'Ağız'
    D = 'En'
    E = 'Körük'
    F = 'Boy'
    
    
'''    

import numpy as np

# Column assignments
J1 = 'Bıçak Bitmiş Ebadı En'
I1 = 'Bıçak Bitmiş Ebadı Boy'
H1 = 'Kağıt Gramajı'
G1 = 'Ağız'
D1 = 'En'
E1 = 'Körük'
F1 = 'Boy'

zenbo_luks_necessary_checked['Kağıt Gramajı.3'] = np.where((zenbo_luks_dropped1[H1] >= 120) & (zenbo_luks_dropped1[H1] <= 250), 1, 0)

zenbo_luks_necessary_checked['Çanta En.2'] = np.where((zenbo_luks_dropped1[D1] >= 18) & (zenbo_luks_dropped1[D1] <= 43), 1, 0)

zenbo_luks_necessary_checked['Körük Açık Çanta Yüksekliği'] = np.where(((zenbo_luks_dropped1[E1] / 2) + zenbo_luks_dropped1[F1] + 1.5 >= 23) & ((zenbo_luks_dropped1[E1] / 2) + zenbo_luks_dropped1[F1] + 1.5 <= 50), 1, 0)

zenbo_luks_necessary_checked['Körük.3'] = np.where((zenbo_luks_dropped1[E1] >= 8) & (zenbo_luks_dropped1[E1] <= 17), 1, 0)

zenbo_luks_necessary_checked['Dip Takviye En'] = np.where(((zenbo_luks_dropped1[D1] - 0.5) >= 17.4) & ((zenbo_luks_dropped1[D1] - 0.5) <= 42.4), 1, 0)

zenbo_luks_necessary_checked['Dip Takviye Boy'] = np.where(((zenbo_luks_dropped1[E1] - 0.5) >= 7.4) & ((zenbo_luks_dropped1[E1] - 0.5) <= 16.4), 1, 0)

zenbo_luks_necessary_checked['Körük Açık Ağız ve Körük Son Uç Arası'] = np.where(zenbo_luks_dropped1[F1] - (zenbo_luks_dropped1[E1] + (zenbo_luks_dropped1[E1] / 2 + 1.5)) >= 4, 1, 0)

zenbo_luks_necessary_checked['Körük Açık Katsız Kısım'] = np.where(zenbo_luks_dropped1[D1] - (2 * (zenbo_luks_dropped1[E1] / 2 + 1.5)) >= 5, 1, 0)


# Check if any of the specified columns have a value of 0
zero_check = ((zenbo_luks_necessary_checked['Kağıt Gramajı.3'] == 0) |
              (zenbo_luks_necessary_checked['Çanta En.2'] == 0) |
              (zenbo_luks_necessary_checked['Körük Açık Çanta Yüksekliği'] == 0) |
              (zenbo_luks_necessary_checked['Körük.3'] == 0) |
              (zenbo_luks_necessary_checked['Dip Takviye En'] == 0) |
              (zenbo_luks_necessary_checked['Dip Takviye Boy'] == 0) |
              (zenbo_luks_necessary_checked['Körük Açık Ağız ve Körük Son Uç Arası'] == 0) |
              (zenbo_luks_necessary_checked['Körük Açık Katsız Kısım'] == 0)) 

# Calculate the 'ZB 1200CT-430' column with the conditions
zenbo_luks_necessary_checked['ZB 60S\n(Soğuk Tutkal)'] = np.where(zero_check, 0, 1)
```
